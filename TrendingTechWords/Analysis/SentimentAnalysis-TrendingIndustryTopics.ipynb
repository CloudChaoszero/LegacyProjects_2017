```python
import sys
sys.path.insert(0, '../resources/')
import csv

import tweepy
import json
import pandas as pd
from access import access_token,access_secret,consumer_key,consumer_secret
```

```python
#Create listof hyped terms
hyped_emergingTech_list = ["Smart Dust","Artificial Intelligence","AI","Quantum Computing","Augmented Reality", "AR","VR",\
                           "IoT","Deep Learning","Data Science","Machine Learning","ML","Autonomous Vehicle","Blockchain","Augmented Reality"\
                          ,"Virtual Reality","Drones","Cybersecurity","Big Data","Analytics", "SaaS", "CRM", "Cloud"]
```

```python
# Take the created list of hype words and edit them for export
for i, hyped_word in enumerate(hyped_emergingTech_list):
    hyped_emergingTech_list.remove(hyped_word)
    new_word = "#"+hyped_word
    hyped_emergingTech_list.insert(i,new_word.replace(" ","") )
```

```python
#Export hyped words to a csv
with open("../Data/trendTechList.csv","w",newline='') as output_csvfile:
    techWords_writer = csv.writer(output_csvfile,delimiter = "\n")
    techWords_writer.writerow(hyped_emergingTech_list)
```

```python
#Initiate Tweepy Setup for Data Collection
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token,access_secret)
api = tweepy.API(auth_handler=auth, parser=tweepy.parsers.JSONParser(),wait_on_rate_limit=True)
```

```python
#Gather 1000 tweets per trending word
trendWord_dict = {}
for word in hyped_emergingTech_list:
    noHash = word.replace("#","")
    searchList = []
    searchList.append(api.search(word, count = 1000) )
    trendWord_dict[noHash] = searchList
```

```python
with open("../Data/trendingWords_twitter.txt","w") as trendingWordsJson:
    json.dump(trendWord_dict, trendingWordsJson)
```

```python
# word Dataframe. For every dataframe, have certain columns. For these certain columns, each row is a tweet
words_dfs = {}
colsOfInterest = ["created_at","favorite_count","retweet_count","text","location"]
for tWord, vals in trendWord_dict.items():
    df = {colsOfInterest[0]: [],colsOfInterest[1]: [],colsOfInterest[2]: [],colsOfInterest[3]: [],colsOfInterest[4]: []}
    statuses = vals[0]["statuses"]
    for status in statuses:
        for c in colsOfInterest:
            try:
                df[c].append(status[c])
            except KeyError:
                df[c].append("No Location")
    data = pd.DataFrame(df)
    data.to_csv("../Data/Results/TrendingWord-%s-Twitter2018.csv" % (tWord),sep = ',')
    words_dfs[tWord] = data
```
