## Investment

### Definitions

1. **Default:** Default is failure to pay interest or principal on a loan or security when due.

    a. **Sovereign Default (National Default)**
    
    b. **Technical Default:** result of not meeting a condition of the loan, and has nothing to do with missing a scheduled loan payment.
    
    c. **Debit Service Default:** Missing a schedule payment on a loan
    
    d. **Strategic Default:** a borrower purposely refuses to make a payment on a loan despite their ability to make the payments.


2. **Generally Accepted Accounting Principles:**
    
    a. **Cost Principle:** the actual cost of assets must be recorded instead of recording the cost based on market value or inflation adjustment.
    
    b. **Revenue Principle:** revenue should be recorded at the time that it is earned, not at the time when payment is received. This prevents errors in accounting caused by delayed payments since any money still owed to the company is evident within the accounting ledger. 
    
    c. **Matching Principle:** expenses should be matched to the revenue they are related to. The expenses are not recorded at the time they are generated, but instead are recorded once they make a contribution to revenue. This allows the profitability of goods and services to be easily evaluated and also illustrates the connection between expense and income, since products and services are directly matched with the revenue they generate.
    
    d. **Disclosure Principle:** all of the financial information disclosed by a business should be released in a form that is easy to understand and that this disclosure should be balanced against the cost of compiling and releasing the information. Any information needed to understand financial statements should be included in the body of the statements, in footnotes or in supplemental documents that are provided alongside the statements. The amount of information disclosed should be sufficient for corporate executives to make decisions regarding the company; unnecessary information should be streamlined to keep the cost of producing the statements down.

3. **Liquidity Ratio:** Measure a company's ability to pay debt obligations and its margin of safety through the calculation of metrics including the current ratio, quick ratio, and operating cash flow ratio. Current liabilities are analyzed in relation to liquid assets to evaluate the coverage of short-term debts in an emergency.
    
    a. solvency: relates to a company's overall ability to pay debt obligations and continue business operations. More total assets than total liabilities to be considered solvent.
    
    b. liquidity: Degree to which an asset or security can be quickly bought or sold in market without affecting asset's price. More current assets than current liabilities to be considered liquid.

4. **Chapter 7 Bankrupty:** Common form of bankruptcy where debts are forgiven, and a variety of personal assets are liquidated to repay remaining debt as much as possible.

5. **Chapter 11 Bankrupty:** Debitors are restructured in a way that debt repyayment become more achievable.

6. **Volatility:** Statistical measure of dispersion of returns for a given security or market index.

7. **Credit Risk:** Risk of default on a debt that may arise from a borrower failing to make required payments.

    a. Credit Default Risk: The risk of loss arising from a debtor being unlikely to pay its loan obligations in full or the debtor is more than 90 days past due on any material credit obligation.
    
    b. Concentration Risk: The risk associated with any single exposure or group of exposures with the potential to produce enough losses to threaten a bank's core operations.
    
    c. Country (Soverign) Risk: The risk of loss arising from a soverign state freezing foreign currency payments (transfer/conversion risk) when it defaults on obligations.

8. **Financial Ratio:** Useful indicators of a firm's performance and financial situation.

    a. **Liquidity Ratio:** Indicator of firm's ability to meet its short-term financial obligations
    
    ${\text{Current Ratio}} = \dfrac{\text{Current Assets}}{\text{Current Liabilities}}$
        
    ${\text{Quick Ratio}} = \dfrac{\text{Current Assets}- \text{Inventory}}{\text{Current Liabilities}}$
        
    ${\text{Cash Ratio}} = \dfrac{\text{Cash} + \text{Marketable Securities}}{\text{Current Liabilities}}$
        
    b. Asset Turnover Ratio: Indicator of how efficiently a firm utilizes its assets.
        
    ${\text{Receivables Turnover}} = \dfrac{\text{Annual Credit Sales}}{\text{Accounts Receivable}}$
        
    ${\text{Average Collection Period}} = \dfrac{\text{Accounts Receivable}}{\text{Annual Credit Sales/365}}$
        
    ${\text{Average Collection Period}} = \dfrac{\text{365}}{\text{Receivables Turnover}}$
        
    ${\text{Inventory Turnover}} = \dfrac{\text{Cost of Goods Sold}}{\text{Average Inventory}}$
        
    ${\text{Inventory Period}} = \dfrac{\text{Average Inventory}}{\text{Annual Cost of Goods Sold/365}}$
        
    ${\text{Inventory Period}} = \dfrac{\text{365}}{\text{Inventory Turnover}}$
    
    c. **Financial Leverage Ratio:** Indicator of the long-term solvency of a firm
    
    ${\text{Debt Ratio}} = \dfrac{\text{Total Debt}}{\text{Total Assets}}$
        
    ${\text{Debt-to-Equity Ratio}} = \dfrac{\text{Total Debt}}{\text{Total Equity}}$
        
    ${\text{Interest Coverage}} = \dfrac{\text{EBIT(Earnings Before Interest and Taxes)}}{\text{Interest Charges}}$
    
    d. **Profitability Ratio:** Measure of success of a firm in profits
    
    ${\text{Gross Profit Margin}} = \dfrac{\text{sales}-\text{Cost of Goods Sold}}{\text{Sales}}$
        
    ${\text{Return on Assets}} = \dfrac{\text{Net Income}}{\text{Total Assets}}$
        
    ${\text{Return on Equity}} = \dfrac{\text{Net Income}}{\text{ Shareholder Equity}}$
    
    e. **Divident Policy Ratio:** Provide insight into divident policy of a firm and the prospects of future growth
    
    ${\text{Divident Yield}} = \dfrac{\text{Dividends Per Share}}{\text{Share Price}}$
        
    ${\text{Payout Ratio}} = \dfrac{\text{Dividends Per Share}}{\text{Earnings Per Share}}$

9. $\beta$: Measure of systematic risk (volatility) 

10. **Bond:** Debt Investment in which an investor loans money to an entity which borrows the funds for a defined period of time.

11. **Accrual Accounting:** Accounting where revenues are reported on the income statement when they are earned. Expenses are matched with the related revenues and/or reported when the expense occurs, not when the cash is paid. The result of accrual accounting is an income statement that better measures the profitability of a company during a specific time period.

12. **General Ledger:** Set of numbered accounts for its accounting records. The ledger provides a complete record of financial transactions over the life of the company.

13. **Stock:** Type of security that signifies ownership in a corporation and represents a cliam on part of the corporation's assets and earnings.

14. **Security:** Fungible, negotiable financial instrument that holds some type of monetary value. It represents an ownership position in a publicly-traded corporation.

15. **Coupon (Interest rate):** Yield paid by a fixed-income security.

16. **Capital Structure:** How a firm finances its overall operations and growth by using different sources of funds. 

17. **Credit Default Swaps (CDS):** Particular type of swap designed to transfer the credit exposure of fixed income products between two or more parties.

18. **Collaterized Debit Obligation (CDO):** Structured asset-backed security that can be thought of as a promise to pay investors in prescrived sequence, based on cash flow the CDO collects from the pool of bonds or other assets it owns.

19. **Levy:** Collection of money owned through financial accounts. Creditors freeze financial account, then take money out of current and future deposits until debt has been covered. Thereafter, freeze will be lifted.

20. **Garnishment:** Collection of money through wages. Creditor forces employer to take out a portion of funds per paycheck until debt has been covered.

21. **Lien:** A lien is a legal right granted by the owner of property, by a law or otherwise acquired by a creditor. A lien serves to guarantee an underlying obligation, such as the repayment of a loan. If the underlying obligation is not satisfied, the creditor may be able to seize the asset that is the subject of the lien

22. **Merger:** a deal to unite two existing companies into one new company

    a. Conglomerate Merger: nothing in common for united companies
    
    b. Horizontal Merger: both companies are in same industry, deal is part of consolidation
    
    c. Market Extension Merger: Companies sell same products but compete in different markets
    
    d. Product Extension Merger: add together prodcuts that go well together
    
    e. Vertical Merger: two companies that make parts for a finished good combine

23. **Acquisitions:** Corporate action in which a company buys most, if not all, of another firm's ownership stakes to assume control of it.

24. **Market Capitalization:** Total dollar market value of a company's outstanding shares

    a. Calculated by "market cap," where market cap is multiplying a company's shares outstanding by the current market price of one share

25. **Equity:** Value of asset less the amount of all liabilities on that asset. $Assets-Liabilities = Equity$

26. **Debitor:** Person or enterprise that owes money to another party

27.  **Creditor:** Person, bank, or enterprise that has lent money or extended credit to another party

28. **Market Index:** Aggregate value produced by combining several stocks or other investment vehicles together and expressing their total values against a base value.

29. **Benchmark:** Standard against which the performance of a security, mutual fund, or investmeng manager can be measured.

30. **Principal:** Original sum of money 

31. **Interest:** Payment from a borrower or deposit-taking financial institution to a lender or depositior.

32. **Cash Flows:** Net amount of cash and cash-equivalents moving into and out of a business.

33. **Dividend:** Distribution of a portion of a company's earnings, decided by board of directors, to a class of its shareholders

34. **Yield:** Income return on an investment, such as interest or dividends received from holding a particular security.

35. **Discount:** refers to the condition of the price of a bond that is lower than par, or face value. $\text{Discount = price paid for security - security's par value}$

36. **Option:** Financial derivative that represents a contract sold by one party to another.

37. **Economies of Scale:** cost advantage that arises with increased output of a product.

38. **Goodwill:** Intangible asset that arises as a result of the acquisition of one company by another for a premium value. Examples of goodwill are company name, patent, technology, etc.

39. **Account Receivable:** refers to outstanding invoices a company has or the money the company is owed from its clients

40. **Account Payable:** money owed by a business to its suppliers shown as a liability on a company's balance sheet.

41. **Deferred Tax liabilities:** Tax liability of the company for the current accounting year

42. **Tangible Assets:** The fixed assets that can be seen or touched.

43. **Intangible Assets:** The non-physical assets of the company.

44. **Non-Current Investment:** The investment whose value will be realised after a specificied period

45. **Gross Block:** It is the gross block of fixed asset, from which accumulated depreciation is deducted to arise at the netblock of the asset.

**Define the Weighted Average Cost of Capital**

Rate that a company is expected to pay on average to all its security holders to finance its assets. It represents the min. return that a company must earn on an existing asset base to satisfy its creditors, owners, and other providers of capital, or they will invest elsewhere.

### Formulas

#### Economics

##### Net Present Value

$NPV = {CF}_0 + \dfrac{{CF}_1}{(1+r)} + \dfrac{{CF}_2}{(1+r)^2} + \dfrac{{CF}_3}{(1+r)^3} + ...+\dfrac{{CF}_N}{(1+r)^N}$

where CF:= Cash Flow

r:= Discount Rate

N:= number of time periods

##### Internal Rate of Return

$NPV = 0 = {CF}_0 + \dfrac{{CF}_1}{(1+r)} + \dfrac{{CF}_2}{(1+r)^2} + \dfrac{{CF}_3}{(1+r)^3} + ...+\dfrac{{CF}_N}{(1+r)^N}$

where CF:= Cash Flow

r:= Discount Rate

N:= number of time periods


##### Perpetuity

$PV = \dfrac{CF}{r}$

where

CF := Cash flow

r:= Discount Rate

##### Growing Perpetuity

${PV}_0 = \dfrac{{CF}_1}{r-g} \text{, } r>g$


where 

CF := Cash flow

r:= Discount Rate

g:= growth rate

###### Weighted Average Cost of Capital

${WACC} = (\dfrac{D}{D+E+P}){r_d} + (\dfrac{E}{D+E+P}){r_e} + (\dfrac{P}{D+E+P}){r_p}$

where

D:= Debt

E:= Equity

P:= Profit

r:= Discount Rate

##### Tax-Adjusted Weighted Average Cost of Capital

$WACC = (\dfrac{D}{D+E})(1- T_c ){r_d} + (\dfrac{E}{D+E}{r_e})$


where

D:= Debt

E:= Equity

1-$T_c$:= Tax adjustment for interest expense


##### Preferred Shares

${PS}_0 = \dfrac{D_p}{r_p}$

where
D:= Dividend


##### Growing Dividend

$P_0 = \dfrac{D_1}{r-g}$


##### Basic Dividend

$P_0 = \dfrac{D_1}{r}$

##### Multiple Growth Rate Dividend

$P_0 = \dfrac{D_0 (1+g_1)^1}{(1+r)^1}+\dfrac{D_0 (1+g_1)^2}{(1+r)^2} +...+\dfrac{D_0 (1+g_1)^N}{(1+r)^N}+[\dfrac{1}{(1+r)^N}\cdot \dfrac{D_{N+1}}{r-g_2}] $

##### Total Percentage return

$\text{Total Percentage Return}:= \dfrac{\text{total dollar return}}{\text{initial investment}}$

#####  Portfolio Expected Return

$E(R_p) = w_1{E(R_1)} +  w_2{E(R_2)} +...+ w_n{E(R_n)}$


$w_i$:= weight

R:= return

where i in [1,n]



##### Portfolio Beta

${\beta}_p = w_1{{\beta}_1} + w_2{{\beta}_2} + ...+ w_n{{\beta}_n}$

where $\beta$:= systematic risk

#####  Security Market Line/CAPM

$E(R_i) = R_f + \beta{[E(R_m)- R_f]}$

where 

$R_i$:= risk on investment

$R_f$:= risk free rate

$R_m$:= expected market return

##### CAPM Regression

$(R_i - R_{f}) = \alpha + \beta{[R_m - R_f ]} + \epsilon$ 

where 

$R_i$:= risk on investment

$R_f$:= risk free rate

$R_m$:= expected market return

$\epsilon$:= some error
#####  Put-Call Parity

Payoff on Bond + Payoff on Call = Payoff on stock + payoff on put

##### Binomial Options Model

High Portfolio Value = Low Portfolio Value

Present Value of Total Portfolio Cost = Present Value of Total Portfolio Value

#####  Black Scholes Model

$C = {SN}(d_1)- X{e}^{-n}{N(d_2 )}$

where $d_1 = \dfrac{ln(\dfrac{S}{X}) + (r+\dfrac{\sigma^2}{2})t}{\sigma \sqrt{t}}$


and 

C:= call 

X:= Strike price

${e}^{-n}$:= Discount Rate

S:= Stock price

N(x):= Cumulative distribution function for normal distribution, where N(x)<1.

t:= time of expiration


Another way to see Black Scholes Model is "Call = get - pay"

### Formulas

#### Percent change (g) in GDP (y) b/t period t and t+1

$$ g = \dfrac{y_{t+1} - y_t}{y_t}$$


#### Measuring Growth

##### Traditional

$$g = \dfrac{y_{t+1} - y_t}{y_t}$$

##### Continuous compound growth rates

$$ \gamma = \log(x_{t+1}) - \log(x_t) )$$

growth rates are additive, averages easy to compute, and little difference in $g$ growth rates are small.

#### Production Function

$$Y = A \dot F(K,L)$$

where L is labor, K is capital, and A is some constant

Considering hours of worked $h$ and some $\alpha$ variable, we modify it like so:

$$ Y = AK^{\alpha}(hL)^{1-\alpha}$$

##### GDP in Expenditure Approach

where Y = Nominal GDP, C = Consumption, I = Investment, G = Government Purchases, X = Exports, M = Imports
